# **Plotting angular part of wavefunctions (Spherical harmonics)**

# **Libraries and defining functions**

```python


%matplotlib inline
import matplotlib.pyplot as plt

from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import scipy.integrate as integrate
import scipy as sp

# Increase resolution for retina display
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')



```

```python
def psi_ang(theta,phi,l=0,m=0):

    sphHarm = sp.special.sph_harm(m,l,theta,phi)

    return sphHarm.real
```

# **Plot**

**Giving different values of l and m to the function:**
= (1).png

where l is the azimutal quantum number, m is the magnetic quantum number and $$ P_{lm}(\cos \theta)$$ are the laguerre polynomials as a function of $$\cos \theta$$

```python
theta, phi = np.linspace(0, np.pi, 100), np.linspace(0, 2*np.pi, 100)

theta, phi = np.meshgrid(phi, theta)
```

**Enter different values that you want to the quantum numbers "l" and "m":**

```python
#request values for l and m
el= int(input("Give me a value for l: "))
em= int(input("Give me a value for m: "))
Ylm = psi_ang(theta,phi,l=el,m=em)

#cartesian coordinates
x = np.sin(phi) * np.cos(theta) * abs(Ylm)
y = np.sin(phi) * np.sin(theta) * abs(Ylm)
z = np.cos(phi) * abs(Ylm)

'''Set up the 3D Canvas'''

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

''' Normalize color bar to [0,1] scale'''

fcolors = (Ylm - Ylm.min())/(Ylm.max() - Ylm.min())

'''Make 3D plot of real part of spherical harmonic'''

ax.plot_surface(x, y, z, facecolors=cm.seismic(fcolors), alpha=0.3)


''' Project 3D plot onto 2D planes'''

cset = ax.contour(x, y, z,20, zdir='z',offset = -1, cmap='summer')
cset = ax.contour(x, y, z,20, zdir='y',offset =  1, cmap='winter' )
cset = ax.contour(x, y, z,20, zdir='x',offset = -1, cmap='autumn')


''' Set axes limit to keep aspect ratio 1:1:1 '''

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)

ax.set_facecolor("black")
```
